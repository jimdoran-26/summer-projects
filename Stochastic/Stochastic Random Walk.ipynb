```python
import random
import matplotlib.pyplot as plt
```

```python
count = 0
lis = []

for i in range(0,50):
    toss = random.randint(1,100)
    if toss>50:
        count=count+1
    else:
        count=count-1
    lis.append(count)
#lis
```

```python
plt.plot(lis)
```
