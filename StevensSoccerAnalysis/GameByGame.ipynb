```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read in offensive stats
off = pd.read_csv('OffensiveStats.csv')
off
```

```python
off = off.drop(['Unnamed: 0'], axis=1)
off = off.drop(['Min'], axis=1)
off = off.drop(['Points'], axis=1)
off = off.drop(['GW'], axis=1)
off = off.drop(['Goals'], axis=1)
off = off.drop(['Assists'], axis=1)



def goals_for(address):
    return address.split("-")[0]

def goals_against(address):
    return address.split("-")[1]

off['GF'] = off['Score'].apply(lambda x: f"{goals_for(x)}")
off['GA'] = off['Score'].apply(lambda x: f"{goals_against(x)}")
off['GF']=pd.to_numeric(off['GF'])
off['GA']=pd.to_numeric(off['GA'])

off = off.drop(['Score'], axis=1)


off
```

```python
def result(x):
    if x['GF'] == x['GA']:
        return 'D'
    elif x['GF'] > x['GA']:
        return 'W'
    else:
        return 'L'

off['Result'] = off.apply(result, axis=1)
off
```

```python
offense_shot=off.groupby(['Result']).mean().drop(['Playoff','Conference'],axis=1)
```

```python
#Read in opponent stats
opp = pd.read_csv('OpponentStats.csv')
opp
```

```python
opp = opp.drop(['Unnamed: 0'], axis=1)
opp = opp.drop(['Min'], axis=1)
opp = opp.drop(['Points'], axis=1)
opp = opp.drop(['GW'], axis=1)
opp = opp.drop(['Goals'], axis=1)
opp = opp.drop(['Assists'], axis=1)



def goals_for(address):
    return address.split("-")[0]

def goals_against(address):
    return address.split("-")[1]

opp['GF'] = opp['Score'].apply(lambda x: f"{goals_for(x)}")
opp['GA'] = opp['Score'].apply(lambda x: f"{goals_against(x)}")
opp['GF']=pd.to_numeric(opp['GF'])
opp['GA']=pd.to_numeric(opp['GA'])

opp = opp.drop(['Score'], axis=1)


opp
```

```python
def result(x):
    if x['GF'] == x['GA']:
        return 'D'
    elif x['GF'] > x['GA']:
        return 'W'
    else:
        return 'L'

opp['Result'] = opp.apply(result, axis=1)
opp
```

```python
opponent_shot=opp.groupby(['Result']).mean().drop(['Playoff','Conference'],axis=1)
```
