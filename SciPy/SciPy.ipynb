```python
#interpolation
#new data points within set of known data points
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
x = np.arange(5,20)
y = np.exp(x/3.0)
f = interpolate.interp1d(x,y)
x1=np.arange(6,12)
y1=f(x1)
plt.plot(x,y,'o',x1,y1,'--')
plt.show()
```
