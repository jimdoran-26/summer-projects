```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
gas.dropna()
gas.head()
```

```python
plt.title('gas prices (usd)')
for c in gas:
    if c!= 'Year':
        plt.plot(gas.Year,gas[c], label = c)
plt.xlabel('Year')
plt.ylabel('Price')
plt.xticks(gas.Year[::3])
plt.legend()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('Gas_price_figure.png',dpi = 300)
plt.show()
```
