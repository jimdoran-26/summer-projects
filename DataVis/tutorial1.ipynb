```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random as rd
```

Basic Graph

```python
x = range(1,11)
y = []
for i in x:
    y.append(rd.randint(0,10))
x,y

plt.figure(figsize=(5,3), dpi = 100)

#line 1
#plt.plot(x,y, label = 'random', color = 'y', linewidth =2, marker = '.', markersize=10, markeredgecolor = 'blue', linestyle = '--')
plt.plot(x,y, '^--', label = '2x')
x2=np.arange(0,4,.5)
plt.plot(x2[:6],x2[:6]**2, 'r', label = 'x^2')
plt.plot(x2[4:],x2[4:]**2,'r--')



plt.title('Graph 1', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('range 1-10')
plt.ylabel('random int')

plt.xticks(range(1,11))
plt.yticks(range(1,15,2))

plt.legend()

plt.savefig('mygraph.png',dpi=300)

plt.show()
```

Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')
patterns = ['/','*','O']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(10,10))

plt.show()
```
